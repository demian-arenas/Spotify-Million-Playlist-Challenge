```python
import os
import json
import random

import pandas as pd
```

```python
mpd_slices_dir = os.listdir("../../data/raw/")
len(mpd_slices_dir)
```

```python
slices_tracklist = []
slices_tracks = []

for slice_dir in random.choices(mpd_slices_dir, k=3):
    playlist_in_slice = []
    slice_dir = f"../../data/raw/{slice_dir}"

    file_json_slice = open(slice_dir)
    json_slice = json.load(file_json_slice)

    playlists = json_slice['playlists']
    for playlist in playlists:
        playlist_id = playlist['pid']
        playlist_tracks = playlist['tracks']
        tracks_indexes = [int(f"{playlist_id}{track['pos']}")
                          for track in playlist_tracks]

        del playlist['tracks']
        playlist_in_slice.append(playlist)

        playlist_tracks = pd.DataFrame(
            data=playlist_tracks, columns=playlist_tracks[0].keys(), index=tracks_indexes)
        playlist_tracks['playlist_id'] = playlist_id
        playlist_tracks['track_id'] = tracks_indexes
        slices_tracks.append(playlist_tracks)

    playlist_processed = pd.DataFrame(
        data=playlist_in_slice, columns=playlist_in_slice[0].keys())
    slices_tracklist.append(playlist_processed)
```

```python
all_tracklists = pd.concat(slices_tracklist)
all_tracklists.info()
```

```python
all_tracks = pd.concat(slices_tracks)
all_tracks.info(verbose=2)
```
